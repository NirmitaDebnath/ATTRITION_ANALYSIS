**3.DATA ANALYSIS**

**Converting yes no from Attrition to 1 for yes and 0 for no. The new column name is Atcount**

```python
df['Atcount'] = df['Attrition'].map({'Yes': 1, 'No': 0})
df
```

**Showing number of Attrition distribution**

```python
df.groupby(['Atcount'],as_index=False)['Atcount'].sum()
```

```python
sns.countplot(x='Attrition',data=df,palette='Set1')
```

**Showing employee count in the company on the basis of Gender**

```python
sns.countplot(y='Gender',data=df,palette='Set2')
```

**From the graph we can see that the number of male employee is more**

**There are total 237 Attrition**

```python
df.groupby(['Gender'],as_index=False)['Atcount'].sum()
```

**Among 237 attrition 87 are Female and 150 are Male**

```python
sns.barplot(x='Gender',y='Atcount',data=df.groupby(['Gender'],as_index=False)['Atcount'].sum(),palette='rocket')
```

**The number of male employeee is more so Attrition of male is also more**

```python
sns.countplot(x='Department',data=df,palette='cubehelix')
```

```python
df.groupby(['Department','Gender'],as_index=False)['Atcount'].sum()
```

```python
sns.barplot(x='Department',y='Atcount',data=df.groupby(['Department'],as_index=False)['Atcount'].sum(),palette='YlOrBr')
```

**We can see max Attrition from Research and Development department, also there are max number of employee present**

```python
df.groupby(['EducationField','Gender'],as_index=False)['Atcount'].sum()
```

```python
plt.figure(figsize=(9,5))
sns.barplot(x='EducationField',y='Atcount',data=df.groupby(['EducationField'],as_index=False)['Atcount'].sum(),palette='deep')
```

**We can understand from graph that Attrition is maximum in Life Science then Medical Field**

```python
plt.figure(figsize=(12,5))
sns.countplot( x='Age',data=df,palette='mako')
```

**We see that max workers are in the age range of 30 to 40**

```python
df.groupby(['Age'],as_index=False)['Atcount'].sum()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Age',y='Atcount',data=df.groupby(['Age'],as_index=False)['Atcount'].sum(),palette='Spectral')
```

**Maximum attrition is between the age group of 26 to 33**

```python
sns.barplot(x='EnvironmentSatisfaction',y='Atcount',data=df.groupby(['EnvironmentSatisfaction'],as_index=False)['Atcount'].sum(),palette='magma')
```

**Maximun Aittrition is from employees with lowest environment Satisfaction**

```python
sns.barplot(x='JobLevel',y='Atcount',data=df.groupby(['JobLevel'],as_index=False)['Atcount'].sum(),palette='flare')
```

**Maximun Aittrition is from employees with lower Job Level**

```python
sns.barplot(x='JobSatisfaction',y='Atcount',data=df.groupby(['JobSatisfaction'],as_index=False)['Atcount'].sum(),palette='viridis')
```

```python
sns.barplot(x='PercentSalaryHike',y='Atcount',data=df.groupby(['PercentSalaryHike'],as_index=False)['Atcount'].sum(),palette='light:b')
```

**Employees whose salary hikes are less are leaving**
